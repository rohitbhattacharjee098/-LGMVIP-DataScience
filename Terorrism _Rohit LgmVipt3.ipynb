## Analysing

```python
plt.subplots(figsize=(22,7))
sns.countplot(x='iyear',data=dt,palette='rocket_r')
plt.xticks(rotation=90)
plt.title('Terrorist Attack Each Year')
```

```python
print("Most terrorist activity in(from this graph) : ", dt['iyear'].value_counts().index[0])
```

```python
plt.figure(figsize=(20, 7))
sns.countplot('imonth', data=dt, palette='husl')
plt.title('No. of Terrorist Activities per Month',fontsize=20)
plt.xlabel('Month', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.show()
```

## Analysing the Active Terrorist Groups

```python
dt.gname.value_counts()[1:10]
```

```python
plt.subplots(figsize=(15,5))
sns.countplot(y='gname',data=dt,order=dt.gname.value_counts()[1:10].index,palette='mako')
plt.title("Active Terrorist Groups")
```

## Attack Types

```python
dt.attacktype1_txt.value_counts()
```

```python
plt.subplots(figsize=(10,7))
sns.countplot(x="attacktype1_txt",data=dt,order=dt.attacktype1_txt.value_counts().index)
plt.xticks(rotation=90)
plt.title("Types of Attack")
```

## Target Areas

```python
plt.subplots(figsize=(26,8))
sns.countplot(x="targtype1_txt",data=dt,order=dt.targtype1_txt.value_counts().index)
plt.xticks(rotation=90)
plt.title("Types of Target")
plt.grid()
plt.show()
```

## Effected Regions

```python
plt.figure(figsize=(15,6))
sns.barplot(dt['region_txt'].value_counts()[:10].index,dt['region_txt'].value_counts()[:10].values)
plt.title('No. of Terrorist Activities by Region', fontsize=20)
plt.xlabel('Regions', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.show()
```

## Affected Countries

```python
Affected_areas = dt.country_txt.value_counts()[:5]
Affected_areas
```

```python
plt.subplots(figsize=(20,5))
sns.countplot(x="country_txt",data=dt,order=dt.country_txt.value_counts()[0:5].index)
plt.xticks(rotation=90)
plt.title("Affected Areas")
plt.show()
```

## Injury Counts

```python
plt.subplots(figsize=(15,7))
sns.countplot(x=dt['nwound'],order=dt.nwound.value_counts()[0:5].index,data=dt,palette = 'husl')
plt.ylabel("Count",size=10)
plt.title('Injuries',size=20)
```

## People killed vs Year

```python
Killedpeople = dt[['iyear',"nkill"]].groupby(['iyear']).sum()
fig,axes = plt.subplots(figsize=(25,15))
Killedpeople.plot(kind='bar',alpha=0.7,ax=axes)
plt.ylabel("No of people killed",fontsize=20)
plt.xlabel("Year",fontsize=20)
top_side= axes.spines["top"]
plt.title("People Died",fontsize=20)
```

## Terrorist attacks vs country and cities(accross the world)

```python
fig,axes=plt.subplots(figsize=(15,10),nrows=2,ncols=1)
sns.barplot(x=dt["country_txt"].value_counts()[:10].values,y=dt["country_txt"].value_counts()[:10].index,ax=axes[0]);
axes[0].set_title("Terrorist Attack Per Country")
sns.barplot(x=dt["city"].value_counts()[1:10].values,y=dt["city"].value_counts()[1:10].index,ax=axes[1]);
axes[1].set_title("Terrorist Attack Per City")
fig.tight_layout()
```
