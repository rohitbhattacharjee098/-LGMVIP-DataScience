```python
import cv2
from PIL import Image
from IPython.display import display
```

## Importing , resizing and Check  the Image 

```python
img = cv2.imread('sea_beech_photo.jpg', 1)
img = cv2.resize(img, (720, 480))
```

```python
cv2.imshow("sea_beech_photo", img)
cv2.waitKey(0)
```

## Coverting the image(1)

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(img))
```

## Converting the RGB image to grayscale

```python
greyscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(greyscale))
```

## Inverting image colours

```python
invimg = cv2.bitwise_not(greyscale)
display(Image.fromarray(invimg))
```

## Image Softening 

```python
softimg = cv2.GaussianBlur(invimg, (5, 5), 0, 0)
display(Image.fromarray(softimg))
```

## Inverting the prev image

```python
softinvImg = 255 - softimg
finalSketch = cv2.divide(greyscale, softinvImg, scale=256.0)
display(Image.fromarray(finalSketch))
```

## Comparison of Main Image and Pencil Sketch

```python
display(Image.fromarray(img))
display(Image.fromarray(finalSketch))
```
